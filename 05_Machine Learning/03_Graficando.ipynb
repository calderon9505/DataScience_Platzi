```
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
```

## Grafica en 3D

Al graficarlo en un entorno local (No notebook) se genera una vista interactiva. Muy firme!!!

```
res = 10
X = np.linspace(-5, 5, res)
Y = np.linspace(-5, 5, res)

# Generar matriz de puntos X, Y
X, Y = np.meshgrid(X, Y)

Z = np.sin(X) + 2*np.sin(Y) 

# Gr√°ficar la superficie
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.cool,
                       linewidth=0, antialiased=False)

fig.colorbar(surf)
```

## Curvas de nivel

```
fig2,ax2 = plt.subplots()
level_map = np.linspace(np.min(Z), np.max(Z), res) 

# Conout Plot
# cp = ax2.contour(X, Y, Z, levels=level_map,cmap=cm.cool)
cp = ax2.contourf(X, Y, Z, levels=level_map,cmap=cm.cool)

fig2.colorbar(cp)
ax2.set_title('Curvas de nivel')
# plt.show()
```
