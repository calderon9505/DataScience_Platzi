````python
import matplotlib.pyplot as plt
import numpy as np
````

## BÃ¡sico

The `plot()` function is used to draw points (markers) in a diagram. By default, the plot() function draws a line from point to point.

````python
xpoints = np.array([0, 2, 5, 6])
ypoints = np.array([0, 250, 150, 200])

plt.plot(xpoints, ypoints)
# plt.show()
````

## Default X-Points

````python
plt.plot(ypoints)
````

## Axis y Legends

````python
plt.plot(ypoints, label='Legend')
plt.axis([-1, 4, -50, 300])
plt.legend(loc='upper left')
````

## Format Strings `fmt`

 `marker|line|color`

````python
plt.plot(ypoints, marker='x', linestyle='dashed', color='green')
````

````python
plt.plot(ypoints, 'x--g')
````

## Estilos y mÃºltiples grÃ¡ficos

````python
# markersize=20
# markeredgecolor='red'
# markerfacecolor='green'
# linestyle='dotted'
# color='blue'      color of the line
# linewidth='4.5'
plt.plot(ypoints, marker='^', ms=20, mec='r', mfc='g', ls=':', c='b', lw='2.5')
plt.plot(ypoints+20)
# You can also plot many lines by adding the points for the x- and y-axis 
# for each line in the same plt.plot() function. 
# plt.plot(x1, y1, x2, y2)
````

## TÃ­tulos, ejes y cuadrÃ­cula

````python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title('TÃ­tulo', fontdict = font1, loc = 'right')
plt.xlabel("eje X", fontdict = font2)
plt.ylabel("eje Y", fontdict = font2)

# plt.grid()
plt.grid(axis = 'x', color = 'gray', linestyle = '--', linewidth = 1)

plt.plot(ypoints)
````

## Subplot

````python
#plot 1:
plt.subplot(1, 2, 1)
plt.plot(xpoints,ypoints)
plt.title('grÃ¡fico 1')

#plot 2:
plt.subplot(1, 2, 2)
plt.plot(xpoints+2,ypoints+100)
plt.title('grÃ¡fico 2')

plt.suptitle('SUPER TÃTULO')

# subplot(row, column, current)
````

A figure with a single Axes

````python
fig, ax = plt.subplots()
ax.plot(xpoints, ypoints)
ax.set_xlabel('x label')
ax.set_xlabel('y label')
ax.set_title('TÃ­tulo')
````

A figure with two Axes

````python
# fig, (ax1, ax2)  = plt.subplots(1, 2, sharey='row')
fig, axs  = plt.subplots(1, 2, sharey='row')

# ax1.plot(xpoints, ypoints)
# ax2.plot(xpoints+2, ypoints+100)

axs[0].plot(xpoints, ypoints)
axs[1].plot(xpoints+2, ypoints+100)

fig.suptitle('Sharing Y axis')
````
