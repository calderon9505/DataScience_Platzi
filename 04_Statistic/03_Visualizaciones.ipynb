# Diagramas de dispersi√≥n

```
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```
# Scatterplot. Diagramas de dispersion.
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

http://seaborn.pydata.org/generated/seaborn.jointplot.html

```
# joint plot
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

```
# Box plot
sns.boxplot(data = iris, x='species', y='sepal_length')
```

```
# Barplot
sns.barplot(data=iris, x='species', y='sepal_length')
```

```
# comprobaci√≥n de que barplot escoge la media para graficar
iris.groupby('species')['sepal_length'].mean()
```

```
from numpy import median
sns.barplot(data=iris, x='species', y='sepal_length', estimator=median)
```

```
iris.groupby('species')['sepal_length'].median()
```

```
from numpy import std
sns.barplot(data=iris, x='species', y='sepal_length', estimator=std)
```

```
iris.groupby('species')['sepal_length'].std()
```

```
sns.pairplot(iris, hue='species')
```
